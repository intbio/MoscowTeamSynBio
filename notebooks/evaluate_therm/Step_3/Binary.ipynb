```python
import Bio
from Bio.SeqUtils.ProtParam import ProteinAnalysis
```

```python
data.sequence[20]   = 'MGSSHHHHHHVSKGEELFTGVVPILVELDGDVNGHKFSVRGEGEGDATYGKLTLKFICTTGKLPVPWPTLVTTLSYGVQCFARYPDHMKQHDFFKSAMPEGYVQERTIFFKDDGCYKTRAEVKFEGDTLVNRIELHGIDFKEDGNILGHKLEYNMNSHNVYIMPDKQNNGIKVNFKIRHNIEDGSVQLDHYQQNTPIGDGPVLLPDYHYLHTWSELSKDPNEKRDHMVVPEFIEAAGITLGMDELYK'
```

```python
def extract_descriptors(sequence):
    analyzed_seq = ProteinAnalysis(sequence)


    descriptors = {
        'molecular_weight': analyzed_seq.molecular_weight(),
        'aromaticity': analyzed_seq.aromaticity(),
        'instability_index': analyzed_seq.instability_index(),
        'isoelectric_point': analyzed_seq.isoelectric_point(),
        'gravy': analyzed_seq.gravy(),
        'charge_at_pH7': analyzed_seq.charge_at_pH(pH=7),
        'ss_a-helix': analyzed_seq.secondary_structure_fraction()[0],
        'ss_turn': analyzed_seq.secondary_structure_fraction()[1],
        'ss_b-sheet': analyzed_seq.secondary_structure_fraction()[2]
    }


    aa_composition = analyzed_seq.get_amino_acids_percent()
    for aa, value in aa_composition.items():
        descriptors[f'aa_{aa}'] = value

    return descriptors

data['descriptors'] = data['sequence'].apply(extract_descriptors)
descriptors_df = data['descriptors'].apply(pd.Series)
data = pd.concat([data, descriptors_df], axis=1)
data.head()
```

```python
data = data.drop('descriptors', axis=1)
```

```python
x = data.drop(['name', 'sequence'], axis = 1)
x
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, labels.clas, test_size = 0.25, stratify = labels.clas)
print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
model_cl_grad = GradientBoostingClassifier(n_estimators=150)
model_cl_grad.fit(x_train, y_train)
y_pred_grad = model_cl_grad.predict(x_test)
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred_grad))
```

```python
from sklearn.ensemble import ExtraTreesClassifier
model_cl_extra = ExtraTreesClassifier(n_estimators=150)
model_cl_extra.fit(x_train, y_train)
y_pred_extra = model_cl_extra.predict(x_test)

print(classification_report(y_test, y_pred_extra))
```
